```python
stocks = ['NTAP','TSLA','FB']
start_date='01/01/2010'
end_date='01/01/2018'
```

```python
import numpy as np 

def download_data(stocks):
    data = web.DataReader(stocks, 'morningstar',
                   start_date, end_date)['Close']
    data.columns = stocks
    return data 

def show_data(data): 
    for stock in data.index.levels[0]:   
        data[stock].plot(figsize=(10,8))
    plt.legend(data.index.levels[0])
    plt.show()

# use Natural Log for normalization purposes
def calculate_returns(data): 
    returns = np.log(data/data.shift(1))
    return returns

def plot_daily_returns(returns): 
    for stock in data.index.levels[0]:   
        returns[stock].plot(figsize=(10,8))
    plt.legend(data.index.levels[0])
    plt.show()

def show_statistics(returns):
    for stock in data.index.levels[0]:
        print "%s %f" %(stock,returns[stock].mean()*252)
    
def initialize_weights():
    weights = np.random.random(len(stocks))
```
