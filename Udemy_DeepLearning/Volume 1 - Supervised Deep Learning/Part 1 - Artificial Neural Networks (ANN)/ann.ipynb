## Data Preprocessing

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# convert France, Geography, Spain -> numerical
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
```

```python
X
```

```python
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_1.fit_transform(X[:, 2])
```

```python
# convert female/male -> numerical
X 
```

```python
# convert numerical value to binary
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X = X[:,1:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Create ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
# Initialising the ANN 
# neural network as classifier 
classifier = Sequential()
```

```python
# Adding input layer and first hidden layer
# (11+1)/2=6,input layer = 11 independent variables, output layer = 1 dependent variable
# uniform function to initialize weight by uniform distribution

#classifier.add(Dense(output_dim = 6, init = 'uniform', activation='relu'))
classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform',input_dim=11))
# p is fraction of input units to drop, if we have ten neurons, p = 0.1 -> 10%, 1 neuron will be disabled
# try 0.1, 0.2, etc. however anything >= 0.5 might have underfitting
# purpose of Dropout is to help revent overfitting 
classifier.add(Dropout(0.1))
```

```python
# Adding second hidden layer
# not necessary, but need more practice to input more layers
# knows from first input layer to expect input_dim = 11, so don't need it 
classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))
classifier.add(Dropout(0.1))
```

```python
# Adding the output layer
# 1 node in output layer so output is 1 
classifier.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
# Compiling the ANN 
# http://ml-cheatsheet.readthedocs.io/en/latest/loss_functions.html
# cross-entropy measures performance of classification model whose output is a probability between 0 and 1 
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fitting classifier to the Training set
# Create your classifier here

# Predicting the Test set results
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

## Make predictions and evaluate the model

```python
# Predicting test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
X_test
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy=round((cm[0][0]+cm[1][1])/(sum(sum(cm))*1.0),2)*100
print "Accuracy: {}%".format(accuracy)
```

## Evaluate, Inprove, and Tune the ANN

```python
# Evaluating the ANN using K-Fold cross validation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
# build ANN classifier
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform',input_dim=11))
    classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))
    classifier.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
# wrap the whole classifier 
# new classifier built on k-fold cross validation
k_fold_classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
accuracies = cross_val_score(estimator=k_fold_classifier, X = X_train, y = y_train, cv = 10, n_jobs=-1) # cv = # of folds for K-fold cv
```

```python
accuracies
```

```python
mean = accuracies.mean()
variance = accuracies.std()
mean
variance
```

```python
# build ANN classifier
def build_tuned_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform',input_dim=11))
    classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))
    classifier.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
# wrap the whole classifier 
# new classifier built on k-fold cross validation
# gs grid search
gs_fold_classifier = KerasClassifier(build_fn = build_tuned_classifier)
parameters = {'batch_size': [25, 32], 
              'epochs': [100, 500], 'optimizer': ['adam','rmsprop']}

grid_search = GridSearchCV(estimator = gs_fold_classifier, 
                           param_grid = parameters, 
                           scoring = 'accuracy', 
                           cv = 10)

grid_search = grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
best_parameters
best_accuracy
```
